# 基于CyeleGAN的军事目标图片风格迁移方法

## 导入数据集并展示原始图片

```python
import matplotlib.pyplot as plt
import os

path='./datasets/tank/testA'
datanames = os.listdir(path)
imgdir = []
img = []
for i in range(len(datanames)):
    imgdir.append(path + '/' + datanames[i])
    img.append(plt.imread(imgdir[i]))

plt.figure(figsize=(15, 15))
plt.subplot(4, 1, 1)
plt.title('dataset')
plt.axis('off')
plt.imshow(img[0])
plt.subplot(4, 1, 2)
plt.axis('off')
plt.imshow(img[1])
plt.subplot(4, 1, 3)
plt.axis('off')
plt.imshow(img[2])
plt.subplot(4, 1, 4)
plt.axis('off')
plt.imshow(img[3])
plt.show()
```
